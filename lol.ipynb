```python
import pandas as pd
import numpy as np

import pandas as pd
import numpy as np

# Создаем данные для df_station
num_stations = 400
station_names = [f'Station_{i}' for i in range(1, num_stations + 1)]
station_ids = np.random.randint(1000, 2000, num_stations)
lines = ['Line_A', 'Line_B', 'Line_C', 'Line_D']
line_ids = np.random.randint(1, 5, num_stations)
metriki = np.random.randn(num_stations)

# Генерируем случайные значения для входов и выходов на станцию
num_entries = np.random.randint(1, 6, num_stations)
num_exits = np.random.randint(1, 6, num_stations)

# Генерируем случайные значения для оборудования на станции
equip_counts = np.random.randint(1, 11, (num_stations, 3))  # Лестницы, двери, эскалаторы

df_station = pd.DataFrame({
    'station': station_names,
    'station_id': station_ids,
    'line': np.random.choice(lines, num_stations),
    'line_id': line_ids,
    'metriki': metriki,
    'input_stair': np.random.randint(1, 11, num_stations),  # Количество лестниц на входе
    'output_stair': np.random.randint(1, 11, num_stations),  # Количество лестниц на выходе
    'count_stair': np.random.randint(1, 11, num_stations),  # Общее количество лестниц
    'input_door': np.random.randint(1, 11, num_stations),  # Количество дверей на входе
    'output_door': np.random.randint(1, 11, num_stations),  # Количество дверей на выходе
    'count_door': np.random.randint(1, 11, num_stations),  # Общее количество дверей
    'input_escalator': np.random.randint(1, 11, num_stations),  # Количество эскалаторов на входе
    'output_escalator': np.random.randint(1, 11, num_stations),  # Количество эскалаторов на выходе
    'count_escalator': np.random.randint(1, 11, num_stations)  # Общее количество эскалаторов
})


# Создаем данные для df_poezdki
num_trips = 2000000
trip_stations = np.random.choice(station_names, num_trips)
trip_station_ids = np.random.choice(station_ids, num_trips)
trip_lines = np.random.choice(lines, num_trips)
trip_line_ids = np.random.choice(line_ids, num_trips)
val_num = np.random.randint(1, 100, num_trips)

# Добавляем столбец с часом отправления
hours = np.random.randint(0, 24, num_trips)

df_poezdki = pd.DataFrame({
    'station': trip_stations,
    'station_id': trip_station_ids,
    'line': trip_lines,
    'line_id': trip_line_ids,
    'num_val': val_num,
    'hour': hours  # Добавляем столбец с часом отправления
})
```

```python
# Генерация временных меток
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2024-01-01')
date_range = pd.date_range(start=start_date, end=end_date, periods=len(df_poezdki))

# Добавление столбца date в датасет df_poezdki
df_poezdki['date'] = date_range
```

```python
df_poezdki1 = df_poezdki.groupby(by=['station','line','hour'])[['num_val','date']].agg({'num_val' : 'sum', 'date' : 'max'})
df_poezdki1=df_poezdki1.reset_index()
df_poezdki1['download_station'] = df_poezdki1.groupby(by=['station','hour'])['num_val'].transform('max')
df_poezdki1['mean_propusk'] = df_poezdki1.groupby(by=['station','hour'])['num_val'].transform('mean')
df_poezdki1['propusk_spos'] = df_poezdki1['download_station']*100/df_poezdki1['mean_propusk']

df_poezdki1
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3, random_state=42, n_init=300)
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
df_poezdki1['station']=le1.fit_transform(df_poezdki1['station'])
le2 = LabelEncoder()
df_poezdki1['line']=le2.fit_transform(df_poezdki1['line'])
le2 = LabelEncoder()
df_poezdki1['date']=le2.fit_transform(df_poezdki1['date'])
cluster= model.fit_predict(df_poezdki1)
df_poezdki1['cluster'] = cluster

df = df_poezdki1.copy()
df['labels'] = model.labels_
df['labels'] = df.labels.astype('category')
pd.set_option('display.max_rows', 80) # для удосбтва анализа делаем
df.groupby('labels', observed=True).agg(['count','mean', 'std']).T
```

```python
df=df_poezdki.copy()
df=df.sort_values(by='date')
df.set_index('date', inplace=True)

# Убедитесь, что данные упорядочены по дате
df = df.asfreq('D').fillna(0)  # Заполнение пропущенных значений нулями
```

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Игнорирование предупреждений, связанных с ARIMA
warnings.filterwarnings("ignore")

# Проверка стационарности временного ряда с помощью теста Дики-Фуллера
result = adfuller(df['num_val'])
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')

# Дифференцирование для получения стационарного временного ряда
df['total_load_diff'] = df['num_val'].diff().dropna()

# Подбор параметров ARIMA (p, d, q) с использованием функции auto_arima
# from pmdarima import auto_arima
# auto_model = auto_arima(df['total_load'], seasonal=False, trace=True)
# print(auto_model.summary())

# Обучение модели ARIMA
model = ARIMA(df['num_val'], order=(5, 1, 2))  # Пример: p=5, d=1, q=2
model_fit = model.fit()
print(model_fit.summary())
```

```python

# Прогнозирование на 2 года (730 дней)
forecast_steps = 365 * 2
forecast = model_fit.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=df.index[-1], periods=forecast_steps+1, inclusive='right')

# Прогнозированные значения и доверительные интервалы
forecast_values = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

# Визуализация
plt.figure(figsize=(12, 6))
plt.plot(df['num_val'], label='Исторические данные')
plt.plot(forecast_index, forecast_values, label='Прогноз')
plt.fill_between(forecast_index, confidence_intervals.iloc[:, 0], confidence_intervals.iloc[:, 1], color='pink', alpha=0.3)
plt.title('Прогноз загруженности станций на 2 года вперед')
plt.xlabel('Дата')
plt.ylabel('Общая загруженность')
plt.legend()
plt.show()
```
